1. Add the dataset's Drive link to your shortcuts in Google Drive.

    Step-by-Step: Adding the Shortcut

    1. Open Google Drive in your web browser.
    2. Go to the "Shared with me" tab on the left-hand sidebar.
    3. Locate the folder (09_challenge_EllisWinterSchool).
    4. Right-click on the folder.
    5. Select "Organize" > "Add shortcut".
    6. In the popup window, select "All locations" > "My Drive".
    7. Click "Add".

2. Then you can use the shortcut path to open the dataset.
